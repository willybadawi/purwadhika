# Step 3: Data Cleaning and Formating

- Hapus data kosong (NaN) di kolom `Income`

Mengapa menghapus data kosong di Kolom `Income`?

Nilai yang hilang di kolom Income dapat menyebabkan hasil analisis tidak akurat atau bias, terutama jika pendapatan merupakan variabel penting dalam analisis.

```python
df.dropna(subset=['Income'], inplace=True)
```

- Periksa kembali apakah masih ada data yang kosong atau tidak.

```python
df.isna().sum()
```

. Merubah tipe data Dt_Customer dari object menjadi `Datetime`

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format = '%d-%m-%Y')
```

```python
df.info()
```

- Merubah tingkat pendidikan '2n Cycle' menjadi `Bacic` dan 'Graduation' menjadi `Bachelor`

```python
df['Education'].unique()
```

```python
df['Education'] = df['Education'].replace('2n Cycle','Basic')
df['Education'] = df['Education'].replace('Graduation','Bachelor')
df['Education'].unique()
```

- Merubah status pernikahan 'Divorced, Widow' menjadi `Married` dan 'Together, Alone, Absurd, YOLO' menjadi `Single`

```python
df['Marital_Status'].unique()
```

```python
df['Marital_Status'] = df['Marital_Status'].replace('Divorced','Married')
df['Marital_Status'] = df['Marital_Status'].replace('Widow','Married')
df['Marital_Status'] = df['Marital_Status'].replace('Together','Single')
df['Marital_Status'] = df['Marital_Status'].replace('Alone','Single')
df['Marital_Status'] = df['Marital_Status'].replace('Absurd','Single')
df['Marital_Status'] = df['Marital_Status'].replace('YOLO','Single')
df['Marital_Status'].unique()
```

```python
df.head()
```

- Menambahkan kolom Age / Usia berdasarkan `Year_Birth`

```python
from datetime import datetime

# Hitung tahun sekarang
current_year = datetime.now().year

# Tambahkan kolom Age berdasarkan Year_Birth
df['Age'] = current_year - df['Year_Birth']

df.head()
```

- Mengkategorikan kelompok usia menjadi 4 kelompok, yaitu :
  - Young_Adults yang berusia dari umur 20 tahun sampai dengan 35 tahun
  - Middle_Aged_Adults yang berusia dari umur 36 tahun sampai dengan 55 tahun
  - Older_Adults yang berusia dari umur 56 tahun sampai dengan 65 tahun
  - Senior_Citizens yang berusia lebih dari umur 66 tahun

```python
def category(Age):
    if 20 <= Age <= 35:
        return 'Young_Adults'
    elif 36 <= Age <= 55:
        return 'Middle_Aged_Adults'
    elif 56 <= Age <= 65:
        return 'Older_Adults'
    elif Age > 66:
        return 'Senior_Citizens'
    
df['Age_Category'] = df['Age'].apply(category)
df.head()
```

- Mencari Outlier pada kolom `Income`

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Misalkan df adalah DataFrame dengan kolom 'Income'
plt.figure(figsize=(8, 6))

# Membuat boxplot untuk kolom 'Income'
sns.boxplot(x=df['Income'])

# Menambahkan judul
plt.title('Boxplot of Income')

# Menampilkan plot
plt.show()
```

```python
import pandas as pd

# Misalkan df adalah DataFrame dengan kolom 'Income'

# Menghitung Q1 (kuartil pertama) dan Q3 (kuartil ketiga)
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)

# Menghitung IQR (Interquartile Range)
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menampilkan batas bawah dan batas atas
print("Lower Bound (Batas Bawah):", lower_bound)
print("Upper Bound (Batas Atas):", upper_bound)

# Menyaring dan menampilkan nilai-nilai outlier
outliers = df[(df['Income'] < lower_bound) | (df['Income'] > upper_bound)]
print("\nOutliers:")
print(outliers['Income'])
```

- Membuang Outlier pada kolom `Income`, sehingga akan terlihat rata-rata pendapatan yang sebenarnya

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Misalkan df adalah DataFrame dengan kolom 'Income'

# Menghitung Q1 (kuartil pertama) dan Q3 (kuartil ketiga)
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)

# Menghitung IQR (Interquartile Range)
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menyaring data untuk menghapus outlier
df1 = df[(df['Income'] >= lower_bound) & (df['Income'] <= upper_bound)]

# Membuat boxplot tanpa outlier
plt.figure(figsize=(8, 6))
sns.boxplot(x=df1['Income'])

# Menambahkan judul
plt.title('Boxplot of Income (Without Outliers)')

# Menghitung rata-rata income tanpa outlier
mean_income_no_outliers = df1['Income'].mean()

# Menampilkan plot
plt.show()
print("Rata-rata Income tanpa Outlier:", mean_income_no_outliers)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Produk yang dianalisis
product_columns = ["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]

# Hitung total pelanggan
total_customers = df1.shape[0]

# Hitung persentase pelanggan yang membeli kembali setiap produk
repurchase_percentages = []

for product in product_columns:
    # Temukan pelanggan yang membeli produk lebih dari sekali
    repurchased_customers = df1[df1[product] > 1]
    repurchase_percentage = (repurchased_customers.shape[0] / total_customers) * 100
    repurchase_percentages.append({'Product': product, 'Repurchase Percentage': repurchase_percentage})

# Konversi ke DataFrame
repurchase_df = pd.DataFrame(repurchase_percentages)

# Buat grafik batang menggunakan Seaborn
plt.figure(figsize=(18, 6))
sns.barplot(x='Product', y='Repurchase Percentage', data=repurchase_df, palette='Blues_d')

# Menambahkan judul dan label
plt.title('Persentase Pelanggan yang Membeli Kembali Produk', fontsize=16, pad=20)
plt.xlabel('Produk', fontsize=12)
plt.ylabel('Persentase (%)', fontsize=12)
plt.ylim(0, 100)

# Tampilkan nilai persentase di atas batang
for index, row in repurchase_df.iterrows():
    plt.text(index, row['Repurchase Percentage'] + 2, f'{row["Repurchase Percentage"]:.2f}%', 
             ha='center', va='bottom', fontsize=12)

plt.xticks(rotation=30, ha='right')

# Mengatur jarak agar tidak terpotong
plt.tight_layout()

# Menyesuaikan jarak antara judul dan grafik
plt.subplots_adjust(top=0.9)  # Menyesuaikan jarak atas agar judul tidak terpotong

plt.show()

```
